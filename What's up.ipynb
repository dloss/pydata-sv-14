```python
repo_root = os.path.expanduser('~/dev/ip/mine')

from subprocess import check_output
```

```python
from datetime import datetime

def commits_between(start, end):
    """count commits between refs"""
    out = check_output(['git', 'log', '--oneline', '%s...%s' % (start, end)], cwd=repo_root)
    return len(out.splitlines())

def commit_date(ref):
    """Get the date of a ref"""
    out = check_output(['git', 'log', '-1', '--format=%at', ref], cwd=repo_root)
    return datetime.fromtimestamp(int(out.decode('utf8')))
```

```python
releases = [
    '0.9',
    '0.10',
    '0.11',
    '0.12',
    '0.13',
    '1.0.0',
    '2.0.0',
]
```

```python
for start, end in zip(releases[:-1], releases[1:]):
    start_tag, end_tag = ('rel-%s' % v for v in (start, end))
    commits = commits_between(start_tag, end_tag)
    start_date = commit_date(start_tag)
    end_date = commit_date(end_tag)
    delta = end_date - start_date
    days = delta.days
    end_date_s = end_date.strftime("%Y-%m-%d")
    per_week = 7.0 * commits / days
    print "{end:5s} ({end_date_s}): {commits:4d} commits in {days} days ({per_week:5.1f} commits/week)".format(**locals())
```

## The dream

IPython is part of a wider Python in Data ecosystem.
But IPython is itself growing into something of an ecosystem:

- IPython core
- IPython.kernel
- IPython.nbformat (basic notebook document implementation)
- IPython.nbconvert (converting notebooks)
- IPython.html (notebook server)
- IPython.qt (qt console)
- IPython.parallel (coordinating multiple kernels)

IPython has been moving fast, so it makes sense to keep all of these things together.
My dream for the next few years is to stabilize IPython enough that it can be broken into several independent repos. Third party adoption of our APIs (especially non-Python kernels) helps motivate some of this stabilization of the layers between IPython components.
