# Static Widgets

In IPython 2.0, widgets are transient.

3.0 will introduce persistence

- Last state in the notebook file
- Many states for static export

```python
%matplotlib inline

import matplotlib.pyplot as plt
import networkx as nx

from IPython.html.widgets import interact

@interact(n=(2,30), p=(0.0, 1.0, 0.001))
def plot_random_graph(n, p):
    g = nx.erdos_renyi_graph(n, p)
    nx.draw(g)
    plt.show()
```

# Multi-user notebook server

<img src="multiuser.svg" width="300px">

Important design points:

- Use existing system accounts
- Multi-user server process spawns notebook servers per user
- Each user's notebook server has full shell permissions
- All of these processes are behind a think HTTP proxy
- Scaling to 10k users is out of scope
